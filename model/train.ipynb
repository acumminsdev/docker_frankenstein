```r
library(caret)
library(ggplot2)
library(pROC)
library(rpart)
library(jsonlite)
library(rpart.plot)
library(biglm)
```

```r
# read in data
df <- read.csv("../data/train.csv")
df$Age[is.na(df$Age)] <- mean(df$Age[!is.na(df$Age)])
df$Pclass <- as.factor(df$Pclass)
```

```r
# column definitions
# sibsp Number of Siblings/Spouses Aboard
# parch Number of Parents/Children Aboard
```

```r
# split data
set.seed(107)
inTrain <- createDataPartition(y = df$Survived, p=0.75, list=F)
df.train <- df[inTrain,]
df.test  <- df[-inTrain,]
```

```r
# fit linear model
fit.lm <- bigglm(Survived ~ Pclass + Sex + Age + SibSp + Parch + Fare, df.train, family = binomial())
 
# fit linear model with interaction variables
fit.lm2 <- bigglm(Survived ~ (Pclass + Sex + Age + SibSp + Fare) ^ 2, df.train, family = binomial())
```

```r
# fit and plot decision tree
fit.dt <- rpart(Survived ~ Pclass + Sex + Age + SibSp + Parch + Fare, data=df.train, method="class")
rpart.plot(fit.dt, type=4, extra=1)
```

```r
# distribution of survived/died versus score
results <- data.frame(Survived = as.factor(df.test$Survived), 
                      predict.lm = predict(fit.lm, df.test, type = "response"),
                      predict.lm2 = predict(fit.lm2, df.test, type = "response"),
                      predict.dt = predict(fit.dt, df.test)[,2]
                      )
```

```r
ggplot(data=results, aes(x = predict.lm, group = Survived, color = Survived)) +
  ggtitle("linear model") +
  geom_density(adjust=1.5)
```

```r
ggplot(data=results, aes(x = predict.lm2, group = Survived, color = Survived)) +
  ggtitle("linear model with 2-way interactions") +
  geom_density(adjust=1.5)
```

```r
ggplot(data=results, aes(x = predict.dt, group = Survived, color = Survived)) +
  ggtitle("decision tree") +
  geom_density(adjust=1.5)
```

```r
# plot ROC & AUC for all models
plot.roc(df.test$Survived, as.numeric(predict(fit.lm, df.test)), print.auc=T)
plot.roc(df.test$Survived, as.numeric(predict(fit.lm2, df.test)), add=T, 
         col="blue", print.auc=T , print.auc.y=0.45)
plot.roc(df.test$Survived,as.numeric(predict(fit.dt, df.test)[,2]), add=T, 
         col="red", print.auc=T, print.auc.y=0.4)
```

```r
model_env <- new.env()

model_env$model <- fit.lm2
model_env$score <- function(model, new_data) {
    new_data$Pclass <- factor(new_data$Pclass, levels = c(1,2,3))
    new_data$Sex <- factor(new_data$Sex, levels = c("female", "male"))
    new_data$Survived <- 0
    predict(model, new_data, type = "response")}
```

```r
saveRDS(model_env, file = "model_env.rds")
```
